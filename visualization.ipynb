```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('Les 2 premiers composants PCA')
targets = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
colors = ['red', 'blue', 'green', 'purple', 'grey', 'orange', 'yellow']
for target, color in zip(targets, colors):
    indice = shealth.concat_data_for_PCA["day"] == target
    ax.scatter(shealth.concat_data_for_PCA.loc[indice, 'PC1'],
               shealth.concat_data_for_PCA.loc[indice, 'PC2'],
               c=color,s=50)
ax.legend(targets)
ax.grid()
plt.show()
```
